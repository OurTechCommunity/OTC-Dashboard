```python
import os
import datetime
import pandas as pd
```

```python
number = []
duration = []
date = []
projects = []
for filename in os.listdir(os.path.join(os.getcwd(), 'data')):
    flag = 0
    project = []
    number.append(filename.split('.')[0])
    with open(os.path.join(os.getcwd(), 'data', filename), 'r', encoding='utf-8') as f:
        for line in f:
            if 'we had a fun' in line: 
                duration.append(line.split('fun ')[1].split(' session')[0])
            elif 'was an amazing' in line:
                duration.append(line.split('amazing ')[1].split(' session')[0])
                
            if "<time:" in line:
                date.append(line.split('<time:')[1].split('T')[0])
            
                
            if "## Projects" in line:
                flag = 1
            if flag==1 and len(line.split('@**'))>1:
                name = line.split('@**')[1].split('**')[0].title()
                if name == 'Sirus':
                    project.append('Darshan Rander')
                elif name == 'Ishandeveloper':
                    project.append('Ishan Sharma')
                else:
                    project.append(name)
        projects.append(', '.join(project))
```

```python
date.pop(14) #because on the 10th index there is a time stamp of another meeting
```

```python
dates_list = [datetime.datetime.strptime(date, '%Y-%m-%d').date() for date in date]
```

```python
final_date = []
for date in dates_list:
    final_date.append((date-datetime.timedelta(days=7)).strftime("%d-%m-%Y"))
```

```python
df = pd.DataFrame({'Number':number, 'Date': final_date, 'Duration': duration, 'Attendees': '', 'Projects': projects})
```

```python
duration_final = []
for i in range(df.shape[0]):
    duration_list = df['Duration'][i].split(' ')
    if len(duration_list)>3:
        duration_final.append(round((int(duration_list[0])*60 + int(duration_list[2]))/60, 2))
    else:
        duration_final.append(round(int(duration_list[0]), 2))
df['Duration'] = duration_final
```
